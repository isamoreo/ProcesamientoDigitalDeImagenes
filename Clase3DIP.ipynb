```
# Display four images
def display4(a, b, c, d, title1="Original", title2="Edited", title3="Edited", title4="Edited"):
    plt.figure(figsize=(12,14))
    
    plt.subplot(141), plt.imshow(a), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(142), plt.imshow(b, cmap='gray'), plt.title(title2)
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(143), plt.imshow(c, cmap='gray'), plt.title(title3)
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(144), plt.imshow(d, cmap='gray'), plt.title(title4)
    plt.xticks([]), plt.yticks([])
    
    plt.show()
```

# **BGR Color Space**

```
#leer uan imagen
# principle of least astonishment (surprise)

img_BGR = cv2.imread('rub07.jpg', cv2.IMREAD_COLOR)

#b,g,r = cv2.split(img_BGR)

# tomar todas las filas y columnas de cada canal, variando el canal
b = img_BGR[:,:,0]
g = img_BGR[:,:,1]
r = img_BGR[:,:,2]

display4(img_BGR, b, g, r, "BGR Original", "B channel", "G channel", "R channel")
```

# **RBG Color Space**

```
# pasar la imagen a RGB
img_RGB = cv2.cvtColor(img_BGR, cv2.COLOR_BGR2RGB)
r,g,b = cv2.split(img_RGB)

display4(img_RGB, r, g, b, "RGB Original", "R channel", "G channel", "B channel")
```

# **No en escala de grises**

```
count = 2
plt.figure(figsize=(12,14))
plt.subplot(1,4,1), plt.imshow(img_RGB/255.0), plt.title("Original RGB")
plt.axis("off")

#for especial, ademas de entregar la posicion me entrega los valores de cada posici√≥n. i toma el valor de la posicion y lab toma el valor dentro de cada posicion(rgb)
for i, lab in enumerate(["R","G","B"]):
  
  #crear una matriz de ceros del mismo tama√±o de la imagen
  crgb = np.zeros(img_RGB.shape)
  
  #se cambia el canar que quiero 
  crgb[:,:,i] = img_RGB[:,:,i]
  
  # Se divide para normalizar y hacer que pueda graficar todos los datos, porque el va a graficar de 0 a 1 y sin dividir se pierde mucha informaci√≥n
  plt.subplot(1,4,count), plt.imshow(crgb/255.0), plt.title(lab)
  plt.axis("off")
  count += 1
```
