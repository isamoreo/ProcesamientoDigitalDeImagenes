```
def displayGray4(a, b, c, d, title1="Original", title2="Edited", title3="Edited", title4="Edited"):
    plt.figure(figsize=(12,20))
    
    plt.subplot(141), plt.imshow(a, cmap='gray'), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(142), plt.imshow(b, cmap='gray'), plt.title(title2)
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(143), plt.plot(c, color='gray'), plt.title(title3)
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(144), plt.plot(d, color='gray'), plt.title(title4)
    plt.xticks([]), plt.yticks([])
  
    plt.show()
    
def displayGray6(a, b, c, d, e, f, title1="Original", title2="Edited", title3="Edited", title4="Edited", title5="Edited", title6="Edited"):
    plt.figure(figsize=(30,4))
    
    plt.subplot(161), plt.imshow(a, cmap='gray'), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(162), plt.imshow(b, cmap='gray'), plt.title(title2)
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(163), plt.imshow(c, cmap='gray'), plt.title(title3)
    plt.xticks([]), plt.yticks([])    
    
    plt.subplot(164), plt.plot(d, color='gray'), plt.title(title4)
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(165), plt.plot(e, color='gray'), plt.title(title5)
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(166), plt.plot(f, color='gray'), plt.title(title6)
    plt.xticks([]), plt.yticks([])

    plt.show()
    
def displayGray8(a, b, c, d, e, f, g, h, title1="Original", title2="Edited", title3="Edited", title4="Edited", title5="Edited", title6="Edited", title7="Edited", title8="Edited"):
    plt.figure(figsize=(30,4))
    
    plt.subplot(181), plt.imshow(a, cmap='gray'), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(182), plt.imshow(b, cmap='gray'), plt.title(title2)
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(183), plt.imshow(c, cmap='gray'), plt.title(title3)
    plt.xticks([]), plt.yticks([])    
    
    plt.subplot(184), plt.imshow(d, cmap='gray'), plt.title(title4)
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(185), plt.plot(e, color='gray'), plt.title(title5)
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(186), plt.plot(f, color='gray'), plt.title(title6)
    plt.xticks([]), plt.yticks([])

    plt.subplot(187), plt.plot(g, color='gray'), plt.title(title7)
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(188), plt.plot(h, color='gray'), plt.title(title8)
    plt.xticks([]), plt.yticks([])
    
    plt.show()
```

# Transformación Logarítmica - oscuro a claro


# 1.

```
import cv2
import numpy as np

#ORIGINAL
img = cv2.imread('dog2.jpg', cv2.IMREAD_GRAYSCALE)

#TRANSFORMADA
#Apply log transform
img_log = (np.log(img+1)/(np.log(1+np.max(img))))*255

#Specify the data type
img_log = np.array(img_log, dtype=np.uint8)

#ECUALIZADA
img_ecu = cv2.equalizeHist(img_log)

#CREAR MASCARA
mask = np.zeros(img.shape[:2], np.uint8)
mask[100:300, 100:400] = 255
masked_img = cv2.bitwise_and(img, img, mask = mask)

#HISTOGRAMAS
hist = cv2.calcHist([img], [0], None, [256], [0, 256])
hist2 = cv2.calcHist([img_log], [0], None, [256], [0, 256])
hist3 = cv2.calcHist([img_ecu], [0], None, [256], [0, 256])
hist4 = cv2.calcHist([img], [0], mask, [256], [0, 256])


displayGray8(img, img_log, img_ecu, masked_img, hist, hist2, hist3, hist4 , 'Original', 'Transformada', 'Ecualizada', 'Mascara', 'Hist Original', 'Hist Transformada', 'Hist Ecualizada', 'Histograma Mascara')
```

```
import numpy as np
import cv2

#Original
img = cv2.imread('dogsmile.jpg',0)
#img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#mas= img.max()
#print(mas)

#Apply Gammma = 2.2 on the normalized image and then multiply by scaling cosnt
img_gamma = np.array(255*(img/255)**2.5, dtype='uint8')

#ECUALIZADA
img_ecu = cv2.equalizeHist(img_gamma)

#CREAR MASCARA
mask = np.zeros(img.shape[:2], np.uint8)
mask[100:300, 100:400] = 255
masked_img = cv2.bitwise_and(img, img, mask = mask)

#HISTOGRAMAS
hist = cv2.calcHist([img], [0], None, [256], [0, 256])
hist2 = cv2.calcHist([img_gamma], [0], None, [256], [0, 256])
hist3 = cv2.calcHist([img_ecu], [0], None, [256], [0, 256])
hist4 = cv2.calcHist([img], [0], mask, [256], [0, 256])

#Display the images in subplots
displayGray8(img, img_gamma, img_ecu, masked_img, hist, hist2, hist3, hist4 , 'Original', 'Transformada Gamma', 'Ecualizada', 'Mascara', 'Hist Original', 'Hist Transformada', 'Hist Ecualizada', 'Histograma Mascara')
```
